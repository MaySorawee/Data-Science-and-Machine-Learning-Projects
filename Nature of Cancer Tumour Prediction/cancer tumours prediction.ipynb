## Split Data

```python
#Split Data to data and target
X = df2.iloc[:,1:]
y = df2["diagnosis"]
print("X shape: ",X.shape)
print("y shape: ",y.shape)
```

## Data scaling

```python
# Using standard scaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)
X_transformed = scaler.transform(X)
X_scaled = pd.DataFrame(X_transformed, columns=["radius_mean","radius_se","radius_worst","concave points_mean"
,"concave points_worst","texture_mean","smoothness_mean","symmetry_mean","fractal_dimension_mean"])
X_scaled.describe()
```

## Outlier Detection

```python
name = ["radius_mean","radius_se","radius_worst","concave points_mean"
,"concave points_worst","texture_mean","smoothness_mean","symmetry_mean","fractal_dimension_mean"]

for i in range(8):
  plt.scatter(X_scaled.iloc[:,i], X_scaled.iloc[:,i+1])
  plt.xlabel(name[i])
  plt.ylabel(name[i+1])
  plt.show()
```

```python
from sklearn.ensemble import IsolationForest

X_outlier = []

for i in range(8):
  X1 = X_transformed[:,i:i+2]

  # Initialize IsolationForest
  iso_forest = IsolationForest(contamination=0.03, random_state=42)  # contamination is the proportion of outliers
  iso_forest.fit(X1)

  # Predict outliers
  outlier_predictions = iso_forest.predict(X1)

  # IsolationForest returns -1 for outliers and 1 for inliers
  outliers = X1[outlier_predictions == -1]
  inliers = X1[outlier_predictions == 1]

  # Get outlier index
  for j in range(len(outlier_predictions)):
    if outlier_predictions[j] == -1:
      X_outlier.append(j)

  # Plotting the data points and highlighting outliers
  plt.figure()
  plt.scatter(inliers[:, 0], inliers[:, 1])
  plt.scatter(outliers[:, 0], outliers[:, 1])
  plt.title('Outlier Detection using Isolation Forest')
  plt.xlabel(name[i])
  plt.ylabel(name[i+1])
  plt.legend()
  plt.show()
```

```python
X_outlier = list(set(X_outlier))
print("The number of outlers = %d" %len(X_outlier))
```

```python
# Delete outlier
# 569-68 = 501 rows left
X_preprocessed = X_scaled.drop(index=X_outlier)

plt.figure(figsize=(10, 10))
plt.title("After")
plt.boxplot(X_preprocessed)
```

```python
# Delete in target
y_preprocessed = y.drop(index=X_outlier)
print(y_preprocessed.shape)
print(X_preprocessed.shape)
```

# Model

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y_preprocessed, test_size=0.2, random_state=42)
print('X_train:',np.shape(X_train))
print('y_train:',np.shape(y_train))
print('X_test:',np.shape(X_test))
print('y_test:',np.shape(y_test))
```

## KNN

```python
neighbors = range(1, 50, 2)

training_scores = []
test_scores = []

for n_neighbors in neighbors:
    knn = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X_train, y_train)
    training_scores.append(knn.score(X_train, y_train))
    test_scores.append(knn.score(X_test, y_test))

plt.figure()
plt.plot(neighbors, training_scores, label="training scores")
plt.plot(neighbors, test_scores, label="test scores")
plt.ylabel("accuracy")
plt.xlabel("n_neighbors")
plt.legend()
```

```python
print(training_scores)
print(test_scores)
print(training_scores[10])
print(test_scores[10])
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Select the number of n_neighbors and predict X_test
knn = KNeighborsClassifier(n_neighbors=21).fit(X_train, y_train)
y_pred = knn.predict(X_test)

# Assuming y_true contains the true labels and y_pred contains the predicted labels
# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

# Compute Sensitivity and Specificity
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)

# Others evaluation
print(classification_report(y_test, y_pred))
print("Sensitivity: %.4f" %sensitivity)
print("Specificity: %.4f" %specificity)

# Plot confusion matrix
plt.figure(figsize=(4, 3))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('KNN')
plt.show()
```
